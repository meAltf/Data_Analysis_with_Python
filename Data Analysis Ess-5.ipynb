```python
plt.plot(days, temperature)
```

```python
plt.plot(days, temperature)
plt.title('Temperature Over a Week')
plt.xlabel('Day')
plt.ylabel('Temperature (Â°C)')
plt.show()
```

```python
plt.scatter(days, temperature)
plt.title('Temperature Over a Week')
plt.xlabel('Day')
plt.ylabel('Temperature (Â°C)')
plt.show()
```

```python
plt.bar(days, temperature)
plt.title('Temperature Over a Week')
plt.xlabel('Day')
plt.ylabel('Temperature (Â°C)')
plt.show()
```
