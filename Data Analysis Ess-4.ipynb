```python
def categorize(Views):
  if Views>=1000:
    return "Platinum"
  elif Views>=500:
    return "Gold"
  elif Views>=100:
    return "Silver"
  else:
    return "MOYE MOYE!!!"
```

```python
#Adding segment based upon the Views count.. using map
```

```python
df_web['segment'] = df_web['Views'].map(categorize)
```
